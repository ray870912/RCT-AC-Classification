```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
before = pd.read_csv("/Users/ray/Thesis/before_2.csv", index_col=0)
after = pd.read_csv("/Users/ray/Thesis/after_2.csv", index_col=0)
```

```python
plt.figure(figsize=(10,10))
plt.plot(before['top_k_categorical_accuracy'], lw=3, label='train')
plt.plot(before['val_top_k_categorical_accuracy'], lw=3, label='validation')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(loc="lower right")
plt.title('Before Feature Extraction Accuracy')
plt.savefig("before_accuracy.jpg")
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.plot(after['top_k_categorical_accuracy'], lw=3, label='train')
plt.plot(after['val_top_k_categorical_accuracy'], lw=3, label='validation')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(loc="lower right")
plt.title('After Feature Extraction Accuracy')
plt.savefig("after_accuracy.jpg")
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.plot(before['loss'], lw=3, label='train')
plt.plot(before['val_loss'], lw=3, label='validation')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(loc="lower right")
plt.title('Before Feature Extraction Loss')
plt.savefig("before_loss.jpg")
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.plot(after['loss'], lw=3, label='train')
plt.plot(after['val_loss'], lw=3, label='validation')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(loc="lower right")
plt.title('After Feature Extraction Loss')
plt.savefig("after_loss.jpg")
plt.show()
```
