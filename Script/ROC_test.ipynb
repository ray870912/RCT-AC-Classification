```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import auc, roc_curve
```

```python
data = pd.read_csv("/Users/ray/Thesis/ROC_prob_test.csv")
```

```python
fpr, tpr, thereshold = roc_curve(data['TRUE'], data['prob'])
auc0 = auc(fpr, tpr)
```

```python
fpr_05, tpr_05, thereshold_05 = roc_curve(data['TRUE'][0:3], data['prob'][0:3])
auc05 = auc(fpr_05, tpr_05)
```

```python
fpr_610, tpr_610, thereshold_610 = roc_curve(data['TRUE'][3:], data['prob'][3:])
auc610 = auc(fpr_610, tpr_610)
```

```python
print(auc05, auc610)
```

```python
plt.figure(figsize=(5,5))
plt.plot(fpr, tpr, lw=3, label='ROC (auc = %0.2f)' % auc0)
# plt.plot(fpr_05, tpr_05, lw=3, label='ROC (auc = %0.2f)' % auc05)
# plt.plot(fpr_610, tpr_610, lw=3, label='ROC (auc = %0.2f)' % auc610)
plt.plot([0, 1], [0, 1], color='black', lw=3, linestyle='--')
plt.legend(loc="lower right")
plt.savefig("ROC curve.jpg")
plt.show()
```
