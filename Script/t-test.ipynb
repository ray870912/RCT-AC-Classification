```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
data = pd.read_csv('/Users/ray/Thesis/ShoulderAll.csv')

# Encoding
data['sex'] = data['sex'].replace({'F': 0, 'M': 1})
data['group'] = data['group'].replace({'AC': 0, 'RCT': 1, 'HC': 2})

AC = data[data['group'] == 0].drop(columns=['ID','group'])
RCT = data[data['group'] == 1].drop(columns=['ID','group'])

pVal = pd.DataFrame(columns=['colName', 'pVal'])
for col in AC.columns:
    t_stat, p_value = stats.ttest_ind(AC[col], RCT[col])
    
    if p_value <= 0.05:
        pVal = pVal.append({'colName':col,'pVal':p_value}, ignore_index=True)

pVal.to_csv('p_val.csv', index=False)
```
